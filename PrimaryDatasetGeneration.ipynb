```python
import PyPDF2 as pdf
import numpy as np
import pandas as pd
import os
import tabula
import warnings
from datetime import datetime
warnings.filterwarnings('ignore')
```

```python
def createDataframe(filepathdir):
    filename=os.listdir(filepathdir)
    finalDF=pd.DataFrame()
    for filepath in filename:
        file=filepathdir+filepath
        tempdf=tabula.read_pdf(file,stream=True,pages=4)[0]
        
        # DROPPING REDUNDANT COLUMNS AND ROWS WHICH DO NOT MAKE SENSE
        if "Unnnamed: 0" in tempdf:
            tempdf.drop('Unnamed: 0', axis=1, inplace=True)
        tempdf.dropna(thresh=2, axis=0, inplace=True)
        # RENAMING COLUMN
        tempdf.rename({'Region States':'States'}, axis=1, inplace=True)
        
        tempdf['States']=tempdf['States'].replace({'NR UP': 'UP', 'NER Meghalaya': 'Meghalaya',"J&K(UT) &":'J&K'})
        if finalDF.empty:
            finalDF=tempdf
            finalDF=finalDF.set_index('States')
        else:
            tempdf=tempdf.set_index('States')
            finalDF=pd.merge(finalDF,tempdf,left_index=True,right_index=True)
    finalDF = finalDF.loc[:,~finalDF.columns.duplicated()]
    return finalDF
```

```python
dataframe=pd.read_csv("Electricity_data.csv")
```

```python
exploratory_dataset=pd.DataFrame(columns=['States','Region','Latitude','Longitude','Date','Electricity Consumed'])
```

```python
exploratory_dataset
```

```python
columns
```

```python
%%time
for column in columns:
    for index,row in dataframe.iterrows():
        data={'States':row['States'],'Region':row['Region'],'Latitude':row['Latitude'],'Longitude':row['Longitude'],'Date':column,'Electricity Consumed':row[column]}
        exploratory_dataset=exploratory_dataset.append(data,ignore_index=True)
```

```python
%%time
for i,row in exploratory_dataset.iterrows():
    string=str(row['Date'].replace('-','/').replace(' ',''))[0:10]
    exploratory_dataset['Date'][i]=string
```

```python
exploratory_dataset.info()
```

```python
exploratory_dataset['Date']=exploratory_dataset['Date'].replace({'08072019': '08-07-2019','09072019': '09-07-2019','10072019': '10-07-2019','11072019': '11-07-2019','12072019': '12-07-2019','13072019': '13-07-2019','14072019': '14-07-2019'})
```

```python
%%time
def try_parsing_date(text):
    anamaly_date=[]
    for fmt in ('%d/%m/%Y','%m-%d-%Y', '%d-%m-%Y', '%m/%d/%Y','%Y-%m-%d'):
        try:
            return datetime.strptime(text, fmt).date()
        except ValueError:
            if text in anamaly_date:
                print('Anomaly detected at this Date -',text)
                anamaly_date.append(text)
            pass
for index,row in exploratory_dataset.iterrows():
    date=try_parsing_date(str(row['Date']))
    exploratory_dataset['Date'][index]=date
```

```python
exploratory_dataset.insert(5,'Day','NaN')
exploratory_dataset.insert(6,'Type of Day','NaN')
exploratory_dataset.insert(7,'Month','NaN')
exploratory_dataset.insert(8,'Season','NaN')
exploratory_dataset.insert(9,'Period','NaN')
```

```python
%%time
exploratory_dataset['Date']=pd.to_datetime(exploratory_dataset['Date'])
exploratory_dataset['Day']=exploratory_dataset['Date'].dt.day_name()
exploratory_dataset['Month']=exploratory_dataset['Date'].dt.month 

Holidaylist=['2019-01-26','2019-03-04','2019-03-21','2019-04-17','2019-04-19','2019-05-18','2019-06-05','2019-08-12',
             '2019-08-15','2019-08-24','2019-09-10','2019-10-02','2019-10-08','2019-10-27','2019-11-10','2019-11-12',
             '2019-12-25','2020-01-26','2020-02-21','2020-03-10','2020-04-06','2020-04-10','2020-05-07','2020-05-24',
            '2020-08-01','2020-08-12','2020-08-15','2020-08-30','2020-10-02']
def parse_holiday(date,day,index):
    if date in Holidaylist or day =="Saturday" or day =="Sunday":
        exploratory_dataset['Type of Day'][index]="Holiday"
    else:
        exploratory_dataset['Type of Day'][index]="Working"

for i,row in exploratory_dataset.iterrows():
    parse_holiday(str(row['Date']).replace(' 00:00:00',''),str(row['Day']),i)
```

```python
exploratory_dataset=exploratory_dataset.drop_duplicates()
```

```python
exploratory_dataset.info()
```

```python
exploratory_dataset.sort_values('Date',inplace=True,kind='mergesort')
```

```python
exploratory_dataset
```

```python
exploratory_dataset=exploratory_dataset[(exploratory_dataset['Date'] >="2019-01-01") & (exploratory_dataset['Date'] <= "2020-10-11")]
```

```python
def getSeason(month,index):
    if (month > 11 or month <= 3):
        exploratory_dataset['Season'][index]="WINTER"
    elif (month >= 4 or month <= 5):
        exploratory_dataset['Season'][index]="SUMMER"
    elif (month >=6 and month <= 9):
        exploratory_dataset['Season'][index]="MONSOON"
    else:
        exploratory_dataset['Season'][index]="AUTUMN"
```

```python
%%time
for i,row in exploratory_dataset.iterrows():
    getSeason(row['Month'],i)
```

```python
exploratory_dataset.loc[(exploratory_dataset['Date'] <="2020-03-24") & (exploratory_dataset['Date'] >= "2019-01-01"),'Period']='pre-Covid-Lockdown'
exploratory_dataset.loc[(exploratory_dataset['Date'] >="2020-03-25") & (exploratory_dataset['Date'] <= "2020-05-31"),'Period']='Covid-Lockdown'
exploratory_dataset.loc[(exploratory_dataset['Date'] >"2020-05-31") & (exploratory_dataset['Date'] <= "2020-10-11"),'Period']='Covid-Unlock-Lockdown'
```
